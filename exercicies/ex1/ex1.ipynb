Gere uma lista com os tipos (‘types’) dos trabalhos ali presentes.
Dessa lista, precisamos saber quanto de cada tipo há.
Adicionalmente, precisamos saber qual é o que possui maior
quantidade, assim como o de menor quantidade.

```python
all_types_arr: list[str] = [ item[1] for item in file.values ]
types_arr: list[str] = list(set(all_types_arr))
for item in types_arr: print(f"We have {sum(str(value).lower().count(item.lower()) for value in all_types_arr)} {item}s")
    
```

Sobre o campo ‘description’, organize duas pilhas: uma com o
quarto caractere e outra com o décimo-terceiro caractere. Para
cada uma dessas listas, indique qual o caractere mais frequente
e o menos frequente.

```python
descriptions: list[str] = [ item[11] for item in file.values ]
fourth_char: list[str] = [item[3] for item in descriptions if item[3] != " "]
thirteenth_char: list[str] = [item[12] for item in descriptions if item[12] != " "]

print(f"The most frequently charter used on fourth char is {max(set(fourth_char), key=fourth_char.count)}")
print(f"The not most frequently charter used on fourth char is {min(set(fourth_char), key=fourth_char.count)}")
print(f"The most frequently charter used on thirteenth char is {max(set(thirteenth_char), key=thirteenth_char.count)}")
print(f"The not most frequently charter used on thirteenth char is {min(set(thirteenth_char), key=thirteenth_char.count)}")
```

Faça uma função para contar quantos caracteres há nesse
campo ‘description’, em cada registro, e monte uma lista com as
respectivas somas. Apresente uma distribuição desses
quantitativos.

```python
def count_char(arr: list[str]):
    count_char = {}
    for char in arr: count_char[char] = count_char.get(char, 0) + 1
    return count_char

register_of_count = [len(count_char(item)) for item in descriptions]

for i, count in enumerate(register_of_count, 1):
    print(f"There are {count} in description {i}")
```
